```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

us = pd.read_csv('https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-counties.csv')

states = ['New York', 'California', 'Texas', 'Washington']

for s in states:
  us_ = us.query(f"state == '{s}'").reset_index()
  plt.plot(us_['date'], us_['cases'], label=s)
  plt.plot()

plt.title('States and Total Cases')
plt.xlabel('Date')
plt.ylabel('Cases')
plt.legend()
plt.show()

for s in states:
  us_ = us.query(f"state == '{s}'").reset_index()
  plt.plot(us_['date'], us_['deaths'], label=s)
  plt.plot()

plt.title('States and Total Deaths')
plt.xlabel('Date')
plt.ylabel('Deaths')
plt.legend()
plt.show()
```

As of 2021-01-18 there has been a total of 536,287 cases and 26,104 deaths related to Covid-19 in New York City, New York. Observing data from a 30 day period from 2020-12-20 to 2021-01-18, there has been a steady increase in the amount of daily cases and deaths with no sign of the curve flattening.

Comparing the data from 4 different states (New York, California, Texas, and Washington), New York had the most cases at the beginning of the pandemic and Washington had the least. However, newer data shows that California has the highest rate of infection compared to the other 4 states and Washington still has the least amount of cases.

New York has the highest amount of death amongst the 4 states with California right behind. Texas has had a slight increase in the amount of death around the middle of the pandemic. Meanwhile, Washington has the least amount of death but the death rate is slowly increasing.

In conclusion, Covid-19 cases and deaths have continued to rise since the beginning of the first recorded case and currently there is no sign of the curve flattening.
